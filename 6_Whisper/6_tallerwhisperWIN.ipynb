```python
# Antes de empezar debes tener ffmpeg instalado en tu sistema.
# Puedes revisar si lo tienes instalado ejecutando el siguiente comando en tu terminal (CMD): ffmpeg -versión
# Si no lo tienes instalado, puedes instalarlo en el mismo CMD con esta línea: winget install Gyan.FFmpeg
# Después de eso debes reiniciar tu editor de código para que pueda reconocer ffmpeg y correr este script sin problemas.

import os
import glob

# Esto buscará automáticamente la carpeta bin de FFmpeg instalada por winget
local_app_data = os.environ.get("LOCALAPPDATA", "")
search_pattern = os.path.join(local_app_data, "Microsoft", "WinGet", "Packages", "Gyan.FFmpeg_*", "**", "bin")
found_paths = glob.glob(search_pattern, recursive=True)

# 2. Si la encuentra, inyectarla directamente en la memoria de este script
if found_paths:
    ffmpeg_bin_path = found_paths[0]
    os.environ["PATH"] = ffmpeg_bin_path + os.pathsep + os.environ["PATH"]
    print(r"✅ FFmpeg enlazado con éxito en:")
    print(f"   {ffmpeg_bin_path}")
else:
    print("❌ No se encontró la carpeta de winget automáticamente.")
    print("   Asegúrate de haber reiniciado tu editor de código.")
```

```python
# Importaremos el modelo a nuestro entorno (esto se deberá hacer siempre)
import whisper
```

```python
# Acá especificamos qué modelo: tiny, base, small, medium, large o turbo
model = whisper.load_model("medium")
# Acá cargamos nuestro audio para ser transcrito
# Debemos tener en cuenta que el nombre del archivo y la dirección de la ruta
# NO tenga espacios, usar "_" guión bajo
# Se permiten audios en .mp3, .wav o .flac como formatos de audio válidos
result = model.transcribe("C:/[PATH]/Whisper_Explained.mp3")
print(result["text"])
```

```python
with open('transcription1.txt', 'w') as f:
    f.write(result["text"])
print("Transcription saved to transcription1.txt")
```

```python
# Hagamos otro ejemplo con un audio más largo (tomará considerablemente más tiempo)
model = whisper.load_model("medium")
result = model.transcribe("C:/[PATH]/API_IBM.mp3")
print(result["text"])
```

```python
# Recuerda cambiar el nombre del archivo resultante, pues al Python ser dinámico, si usas el mismo 
# nombre se borrará la transcripción anterior
with open('transcription2.txt', 'w') as f:
    f.write(result["text"])
print("Transcription saved to transcription2.txt")
```
