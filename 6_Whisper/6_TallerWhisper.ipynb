```python
# Importaremos el modelo a nuestro entorno (esto se deberá hacer siempre)
import whisper
```

```python
# Acá especificamos qué modelo: tiny, base, small, medium, large o turbo
model = whisper.load_model("medium")
# Acá cargamos nuestro audio para ser transcrito
# Debemos tener en cuenta que el nombre del archivo y la dirección de la ruta
# NO tenga espacios, usar "_" guión bajo
# Se permiten audios en .mp3, .wav o .flac como formatos de audio válidos
result = model.transcribe("~/Whisper_Explained.mp3")
print(result["text"])
```

```python
with open('transcription1.txt', 'w') as f:
    f.write(result["text"])
print("Transcription saved to transcription1.txt")
```

```python
# Hagamos otro ejemplo con un audio más largo (tomará considerablemente más tiempo)
model = whisper.load_model("medium")
result = model.transcribe("~/API_IBM.mp3")
print(result["text"])
```

```python
# Recuerda cambiar el nombre del archivo resultante, pues al Python ser dinámico, si usas el mismo nombre
# se borrará la transcripción anterior
with open('transcription2.txt', 'w') as f:
    f.write(result["text"])
print("Transcription saved to transcription2.txt")
```
